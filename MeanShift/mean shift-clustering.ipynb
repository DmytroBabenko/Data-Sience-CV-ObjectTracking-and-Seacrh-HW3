```python
import numpy as np
from sklearn.cluster import MeanShift, estimate_bandwidth
```

```python
def read_points(file):
    points = []
    with open(file) as fp:  
        line = fp.readline()
        while line:
            splitted = line.lstrip().rstrip().split('    ')
            x = int(splitted[0])
            y = int(splitted[1])
            points.append([x, y])
            line = fp.readline()
            
    return points
```

```python
import matplotlib.pyplot as plt
def plot_points(points, color, increase_size=True):
    x = [p[0] for p in points]
    y = [p[1] for p in points]
    
    if increase_size:
        fig=plt.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')
    plt.grid()
    plt.title('Points')
    plt.plot(x, y, color)
    plt.show()
```

```python
colors = ['r+', 'g+', 'b+', 'rv', 'gv', 'bv', 'rx', 'gx', 'bx', 'ro', 'c+', 'm+', 'y+', 'c<', 'm<']
def plot_clustered_points(points, cluster_results, increase_size=True):
    assert len(points) == len(cluster_results)
    if increase_size:
        fig=plt.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')
    plt.grid()
    size_colors = len(colors)
    for i in range(0, len(points)):
        plt.plot(points[i][0], points[i][1], colors[cluster_results[i] % size_colors])
        
    plt.title('Mean shift clustering result')
    plt.show()
        
   
    
```

## S1

```python
points = read_points('dataset/s1.txt')
plot_points(points, 'r+')
```

```python
bandwith = estimate_bandwidth(points, quantile=0.05, n_samples=len(points))
bandwith
```

```python
clustering = MeanShift(bandwith).fit(points)
clustering.labels_
```

```python
clustered = clustering.predict(points)
```

```python
plot_clustered_points(points, clustered)
```

```python
def read_1d_points(file):
    points = []
    with open(file) as fp:  
        line = fp.readline()
        while line:
            splitted = line.lstrip().rstrip().split('    ')
            x = int(splitted[0])
            points.append(x)
            line = fp.readline()
            
    return points
```

```python
def plot_1d_points(points, color, increase_size=True):
    if increase_size:
        fig=plt.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')
    plt.grid()
    plt.title('Points')
    x = np.array(points)
    y = np.zeros(len(points))
    plt.plot(x, y, color)
    plt.show()
```
